## The Transformer Architecture (Kiến trúc Transformer)

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Để hiểu được trạng thái hiện tại của các <b>LLM(mô hình ngôn ngữ lớn)</b>, chúng ta phải quay trở lại bài báo <b>"Attention is All You Need"</b> của Google vào năm 2017. Trong bài báo này, kiến trúc <b>Transformer</b> đã được giới thiệu với thế giới, và nó đã thay đổi ngành công nghiệp mãi mãi.</p>

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Mặc dù mạng nơ-ron nhân tạo (neural networks) có thể được sử dụng để cho phép máy tính hiểu văn bản, nhưng các mô hình này cực kỳ hạn chế do thực tế là chúng chỉ cho phép máy xử lý từng từ một, dẫn đến việc mô hình không thể nắm bắt được toàn bộ ngữ cảnh của văn bản.</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Tuy nhiên, <b>kiến trúc Transformer</b> dựa trên cơ chế chú ý <b>(attention mechanism)</b>, cho phép mô hình xử lý toàn bộ câu hoặc đoạn văn cùng một lúc, thay vì từng từ một. Đây là bí mật chính đằng sau khả năng hiểu ngữ cảnh đầy đủ, mang lại sức mạnh lớn hơn cho tất cả các mô hình xử lý ngôn ngữ này.</p>

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Việc xử lý đầu vào văn bản với kiến trúc Transformer dựa trên <mark
          color:white;
          border-radius:4px;
          opacity:1.0"><b>tokenization</b></mark>, là quá trình chuyển đổi văn bản thành các thành phần nhỏ hơn được gọi là token. Đây có thể là từ, từ phụ, ký tự hoặc nhiều thành phần khác.</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Các token sau đó được ánh xạ tới các ID số, là duy nhất cho mỗi từ hoặc từ phụ. Mỗi ID sau đó được chuyển đổi thành một <b>embedding:</b> một vectơ dày đặc, nhiều chiều chứa các giá trị số. Các giá trị này được thiết kế để nắm bắt ý nghĩa ban đầu của các token và đóng vai trò là đầu vào cho mô hình Transformer.</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Điều quan trọng cần lưu ý là các embedding này có nhiều chiều, với mỗi chiều nắm bắt các khía cạnh nhất định về ý nghĩa của một token. Do bản chất nhiều chiều của chúng, embedding không dễ dàng được con người giải thích, nhưng các mô hình Transformer dễ dàng sử dụng chúng để xác định và nhóm các token có ý nghĩa tương tự trong không gian vectơ.</p>

##The Model(Mô hình)

##Evaluation Metrics (Các chỉ số đánh giá)

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Việc đánh giá hiệu suất <b>(evaluation)</b> cho các mô hình ngôn ngữ có thể khá phức tạp, đặc biệt là khi nói đến việc tóm tắt văn bản. Mục tiêu của mô hình của chúng ta là tạo ra một câu ngắn mô tả nội dung của một đoạn hội thoại <b>(dialogue)</b>, đồng thời duy trì tất cả các thông tin quan trọng trong đoạn hội thoại đó.</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Một trong những số liệu định lượng mà chúng ta có thể sử dụng để đánh giá hiệu suất là <b>ROUGE Score</b>.Nó được coi là một trong những số liệu tốt nhất để tóm tắt văn bản và nó đánh giá hiệu suất bằng cách so sánh chất lượng của bản tóm tắt do máy tạo ra <b>(machine-generated summary)</b> với bản tóm tắt do con người tạo ra <b>(human-generated summary)</b> được sử dụng để tham khảo</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Sự tương đồng giữa hai bản tóm tắt được đo bằng cách phân tích các n-gram <b>(n-grams)</b> chồng lên nhau, có thể là các từ đơn lẻ hoặc chuỗi từ có trong cả hai bản tóm tắt. Đây có thể là các unigram (ROUGE-1), trong đó chỉ đo sự chồng chéo của các từ duy nhất; bigram (ROUGE-2), trong đó chúng ta đo sự chồng chéo của các chuỗi hai từ; trigram (ROUGE-3), trong đó chúng ta đo sự chồng chéo của các chuỗi ba từ; Vân vân. Bên cạnh đó, chúng ta cũng có:</p>

<div style = "margin-left: 25px;">
    
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a"><b>• ROUGE-L</b>: Nó đo lường Chuỗi con chung dài nhất <i>(Longest Common Subsequence - LCS)</i> giữa hai bản tóm tắt, giúp nắm bắt phạm vi nội dung của văn bản do máy tạo ra <i>(machine-generated text)</i>. Nếu cả hai bản tóm tắt đều có chuỗi "The banana is delicious", chúng ta sẽ có một kết quả khớp bất kể chúng xuất hiện ở đâu trong cả hai văn bản.</p>
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">
          Nói cách khác, <b>ROUGE-L</b> không quan tâm đến vị trí chính xác của các từ trong câu, mà chỉ quan tâm đến việc liệu chúng có cùng xuất hiện trong cả hai bản tóm tắt hay không. Miễn là chuỗi "the apple is green" được tìm thấy trong cả hai, nó sẽ được coi là một điểm tương đồng và góp phần vào điểm ROUGE-L.</p>   
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a"><b>• ROUGE-S</b>: Nó đánh giá sự chồng chéo (overlap) của các skip-bigram, là các bigram cho phép khoảng trống giữa các từ. Điều này giúp đo lường sự mạch lạc của bản tóm tắt do máy tạo ra . Ví dụ, trong cụm từ "this banana is absolutely delicious", chúng ta tìm thấy một kết quả khớp cho các thuật ngữ như "banana" và "delicious", nếu đó là những gì chúng ta đang tìm kiếm.</p>
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">
          <b>ROUGE-S</b> tập trung vào việc đánh giá xem các cặp từ quan trọng có xuất hiện cùng nhau trong cả hai bản tóm tắt hay không, ngay cả khi chúng bị cách nhau bởi các từ khác. Điều này giúp đánh giá xem bản tóm tắt do máy tạo ra có giữ được các mối quan hệ ngữ nghĩa quan trọng từ văn bản gốc hay không, góp phần vào sự mạch lạc của bản tóm tắt.</p>
</div>

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Các điểm số này thường có thể nằm trong khoảng từ 0 đến 100, trong đó 0 cho biết không có kết quả khớp và 100 cho biết sự khớp hoàn hảo giữa hai bản tóm tắt.</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a"> Bên cạnh các số liệu định lượng <b>(Quantitative Metrics)</b>, việc sử dụng đánh giá của con người <b>(Human Evaluation)</b> để phân tích kết quả đầu ra của các mô hình ngôn ngữ cũng rất hữu ích, vì chúng ta có khả năng hiểu văn bản theo cách mà máy móc không thể. Vì vậy, chúng ta có thể đọc đoạn hội thoại và sau đó đọc bản tóm tắt để kiểm tra xem đó có phải là bản tóm tắt chính xác hay không.</p>

````python
def display_feature_list(features, feature_type):

    '''
    This function displays the features within each list for each type of data
    '''

    print(f"\n{feature_type} Features: ")
    print(', '.join(features) if features else 'None')

def describe_df(df):
    """
    This function prints some basic info on the dataset and
    sets global variables for feature lists.
    """

    global categorical_features, continuous_features, binary_features
    categorical_features = [col for col in df.columns if df[col].dtype == 'object']
    binary_features = [col for col in df.columns if df[col].nunique() <= 2 and df[col].dtype != 'object']
    continuous_features = [col for col in df.columns if df[col].dtype != 'object' and col not in binary_features]

    print(f"\n{type(df).__name__} shape: {df.shape}")
    print(f"\n{df.shape[0]:,.0f} samples")
    print(f"\n{df.shape[1]:,.0f} attributes")
    print(f'\nMissing Data: \n{df.isnull().sum()}')
    print(f'\nDuplicates: {df.duplicated().sum()}')
    print(f'\nData Types: \n{df.dtypes}')

    #negative_valued_features = [col for col in df.columns if (df[col] < 0).any()]
    #print(f'\nFeatures with Negative Values: {", ".join(negative_valued_features) if negative_valued_features else "None"}')

    display_feature_list(categorical_features, 'Categorical')
    display_feature_list(continuous_features, 'Continuous')
    display_feature_list(binary_features, 'Binary')

    print(f'\n{type(df).__name__} Head: \n')
    display(df.head(5))
    print(f'\n{type(df).__name__} Tail: \n')
    display(df.tail(5))
````

````python
def histogram_boxplot(df,hist_color, box_color, height, width, legend, name):
    '''
    This function plots a Histogram and a Box Plot side by side

    Parameters:
    hist_color = The color of the histogram
    box_color = The color of the boxplots
    heigh and width = Image size
    legend = Either to display legend or not
    '''

    features = df.select_dtypes(include = [np.number]).columns.tolist()

    for feat in features:
        try:
            fig = make_subplots(
                rows=1,
                cols=2,
                subplot_titles=["Box Plot", "Histogram"],
                horizontal_spacing=0.2
            )

            density = gaussian_kde(df[feat])
            x_vals = np.linspace(min(df[feat]), max(df[feat]), 200)
            density_vals = density(x_vals)

            fig.add_trace(go.Scatter(x=x_vals, y = density_vals, mode = 'lines',
                                     fill = 'tozeroy', name="Density", line_color=hist_color), row=1, col=2)
            fig.add_trace(go.Box(y=df[feat], name="Box Plot", boxmean=True, line_color=box_color), row=1, col=1)

            fig.update_layout(title={'text': f'<b>{name} Word Count<br><sup><i>&nbsp;&nbsp;&nbsp;&nbsp;{feat}</i></sup></b>',
                                     'x': .025, 'xanchor': 'left'},
                             margin=dict(t=100),
                             showlegend=legend,
                             template = template,
                             #plot_bgcolor=bg_color,paper_bgcolor=paper_color,
                             height=height, width=width
                            )

            fig.update_yaxes(title_text=f"<b>Words</b>", row=1, col=1, showgrid=False)
            fig.update_xaxes(title_text="", row=1, col=1, showgrid=False)

            fig.update_yaxes(title_text="<b>Frequency</b>", row=1, col=2,showgrid=False)
            fig.update_xaxes(title_text=f"<b>Words</b>", row=1, col=2, showgrid=False)

            fig.show()
            print('\n')
        except Exception as e:
            print(f"An error occurred: {e}")
````

````python
def plot_correlation(df, title, subtitle, height, width, font_size):
    '''
    This function is resposible to plot a correlation map among features in the dataset.

    Parameters:
    height = Define height
    width = Define width
    font_size = Define the font size for the annotations
    '''
    corr = np.round(df.corr(numeric_only = True), 2)
    mask = np.triu(np.ones_like(corr, dtype = bool))
    c_mask = np.where(~mask, corr, 100)

    c = []
    for i in c_mask.tolist()[1:]:
        c.append([x for x in i if x != 100])



    fig = ff.create_annotated_heatmap(z=c[::-1],
                                      x=corr.index.tolist()[:-1],
                                      y=corr.columns.tolist()[1:][::-1],
                                      colorscale = colormap)

    fig.update_layout(title = {'text': f"<b>{title} Heatmap<br><sup>&nbsp;&nbsp;&nbsp;&nbsp;<i>{subtitle}</i></sup></b>",
                                'x': .025, 'xanchor': 'left', 'y': .95},
                    margin = dict(t=210, l = 110),
                    yaxis = dict(autorange = 'reversed', showgrid = False),
                    xaxis = dict(showgrid = False),
                    template = template,
                    #plot_bgcolor=bg_color,paper_bgcolor=paper_color,
                    height = height, width = width)


    fig.add_trace(go.Heatmap(z = c[::-1],
                             colorscale = colormap,
                             showscale = True,
                             visible = False))
    fig.data[1].visible = True

    for i in range(len(fig.layout.annotations)):
        fig.layout.annotations[i].font.size = font_size

    fig.show()
````

````python
def compute_tfidf(df_column, ngram_range=(1,1), max_features=15):
    vectorizer = TfidfVectorizer(max_features=max_features, stop_words='english', ngram_range=ngram_range)
    x = vectorizer.fit_transform(df_column.fillna(''))
    df_tfidfvect = pd.DataFrame(x.toarray(), columns=vectorizer.get_feature_names_out())
    return df_tfidfvect
````

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Chúng ta sẽ sử dụng hàm <code>clean_tags</code> được định nghĩa dưới đây để loại bỏ các thẻ này khỏi văn bản, giúp văn bản trở nên sạch hơn.</p>

````python
def clean_tags(text):
    clean = re.compile('<.*?>') # Compiling tags
    clean = re.sub(clean, '', text) # Replacing tags text by an empty string

    # Removing empty dialogues
    clean = '\n'.join([line for line in clean.split('\n') if not re.match('.*:\s*$', line)])

    return clean
````

````python
test1 = clean_tags(train['dialogue'].iloc[14727]) # Applying function to example text
test2 = clean_tags(test['dialogue'].iloc[0]) # Applying function to example text

# Printing results
print(test1)
print('\n' *3)
print(test2)
````

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">
Có thể thấy rằng chúng ta đã loại bỏ thành công các thẻ khỏi văn bản. Bây giờ, chúng ta sẽ định nghĩa hàm <code>clean_df</code>, trong đó sẽ áp dụng <code>clean_tags</code> cho toàn bộ các bộ dữ liệu.</p>

````python
# Defining function to clean every text in the dataset.
def clean_df(df, cols):
    for col in cols:
        df[col] = df[col].fillna('').apply(clean_tags)
    return df

# Cleaning texts in all datasets
train = clean_df(train,['dialogue', 'summary'])
test = clean_df(test,['dialogue', 'summary'])
val = clean_df(val,['dialogue', 'summary'])
````

Chúng ta sẽ đi sâu vào giải thích các bước mà hàm này thực hiện để giúp bạn hiểu rõ hơn về cách nó tiền xử lý dữ liệu.</p>

<div style = "margin-left: 25px;">
    
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a"><b>• <code>inputs = [doc for doc in examples["dialogue"]]:</code></b>
          
- Trong dòng này, chúng ta đang lặp qua từng đoạn hội thoại trong tập dữ liệu và lưu chúng vào danh sách inputs để chuẩn bị làm đầu vào cho mô hình.
- Đây là cách giúp tạo ra một danh sách các đoạn hội thoại cần được xử lý.</p>
    
<p style="font-family: Calibri, serif; text-align: left;

- Dòng này sử dụng <code>tokenizer</code> để chuyển đổi các đoạn hội thoại thành các token, giúp mô hình BART có thể hiểu được dữ liệu đầu vào.
- <code>max_length</code>=1024 đảm bảo rằng độ dài tối đa của mỗi đoạn văn bản là 1024 token. Nếu đoạn văn bản dài hơn, phần dư sẽ bị cắt bỏ nhờ tham số <code>truncation=True</code>.</p>
    
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a"><b>• <code>labels = tokenizer(text_target=examples["summary"], max_length=128, truncation=True):</code></b>
          
- Dòng này thực hiện quá trình tokenization tương tự, nhưng đối với tóm tắt tham chiếu (summaries), thay vì đoạn hội thoại.
- max_length=128 có nghĩa là chúng ta mong muốn tóm tắt có độ dài ngắn hơn nhiều so với đoạn hội thoại, và truncation=True giúp đảm bảo tóm tắt không vượt quá 128 token.
</p>
    
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a"><b>• <code>model_inputs["labels"] = labels["input_ids"]:</code></b>
          
- Dòng này thêm labels (các token của tóm tắt tham chiếu) vào trong model_inputs để chúng trở thành một phần của bộ dữ liệu đã được tiền xử lý.
- Các input_ids từ tokenized labels được gán vào trường labels trong bộ dữ liệu đầu vào, để mô hình có thể học được cách dự đoán tóm tắt từ hội thoại.</p>

````python
def preprocess_function(examples):
    inputs = [doc for doc in examples["dialogue"]]
    model_inputs = tokenizer(inputs, max_length=1024, truncation=True)

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples["summary"], max_length=128, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Applying preprocess_function to the datasets
tokenized_train = train_ds.map(preprocess_function, batched=True,
                               remove_columns=['id', 'dialogue', 'summary', '__index_level_0__']) # Removing features

tokenized_test = test_ds.map(preprocess_function, batched=True,
                               remove_columns=['id', 'dialogue', 'summary']) # Removing features

tokenized_val = val_ds.map(preprocess_function, batched=True,
                               remove_columns=['id', 'dialogue', 'summary']) # Removing features

# Printing results
print('\n' * 3)
print('Preprocessed Training Dataset:\n')
print(tokenized_train)
print('\n' * 2)
print('Preprocessed Test Dataset:\n')
print(tokenized_test)
print('\n' * 2)
print('Preprocessed Validation Dataset:\n')
print(tokenized_val)
````

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Hàm <code>compute_metrics</code> cũng có sẵn trong tài liệu. Trong hàm này, chúng ta chủ yếu trích xuất các bản tóm tắt do mô hình tạo ra, cũng như các bản tóm tắt do con người tạo ra, và giải mã chúng. Sau đó, chúng ta sử dụng ROUGE để so sánh mức độ tương đồng của chúng nhằm đánh giá hiệu suất. </p>

````python
metric = evaluate.load('rouge') # Loading ROUGE Score
````

````python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred# Obtaining predictions and true labels

    # Decoding predictions
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)

    # Obtaining the true labels tokens, while eliminating any possible masked token (i.e., label = -100)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Rouge expects a newline after each sentence
    decoded_preds = ["\n".join(nltk.sent_tokenize(pred.strip())) for pred in decoded_preds]
    decoded_labels = ["\n".join(nltk.sent_tokenize(label.strip())) for label in decoded_labels]


    # Computing rouge score
    result = metric.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)
    result = {key: value.mid.fmeasure * 100 for key, value in result.items()} # Extracting some results

    # Add mean-generated length
    prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in predictions]
    result["gen_len"] = np.mean(prediction_lens)

    return {k: round(v, 4) for k, v in result.items()}
````

Tôi khuyến khích bạn dành thời gian để tìm hiểu <a href = "https://huggingface.co/docs/transformers/main_classes/trainer#transformers.Seq2SeqTrainingArguments">tài liệu</a> của nó để làm quen với các tham số này.</p>

Vì chúng ta đã đặt <code>load_best_model_at_end = True</code> trong các tham số huấn luyện, Trainer tự động lưu mô hình có hiệu suất tốt nhất, trong trường hợp này là mô hình có <code>Validation Loss</code> thấp nhất.</p>

          font-size: 24px; letter-spacing: .85px;color: #02011a">Epoch thứ hai là epoch có validation loss thấp nhất, với giá trị <code><b>1.443861</b></code>. Nó cũng đạt được điểm số cao nhất cho <code>Rouge1</code> và <code>Rouge2</code>, cũng như điểm số <code>Rougelsum</code> cao nhất.</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Tôi chưa trình bày điểm số <code><b>Rougelsum</b></code> trước đó. Theo tài liệu của <a href = "https://pypi.org/project/rouge-score/">the rouge-score library</a>, chúng ta có thể kết luận rằng điểm số này tương tự như điểm số RougeL, nhưng đo lường phạm vi nội dung ở cấp độ câu, thay vì toàn bộ bản tóm tắt.</p>
          
<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Cột <code>Gen Len</code> cho chúng ta biết độ dài trung bình của các bản tóm tắt do mô hình tạo ra. Điều quan trọng là nhớ rằng chúng ta muốn có các văn bản ngắn gọn nhưng đầy đủ thông tin. Trong trường hợp này, epoch thứ hai cũng tạo ra các bản tóm tắt ngắn nhất trung bình.</p>

#Evaluating and Saving Model (Đánh giá và lưu mô hình)

<p style="font-family: Calibri, serif; text-align: left;
          font-size: 24px; letter-spacing: .85px;color: #02011a">Sau khi huấn luyện và kiểm tra mô hình, chúng ta có thể đánh giá hiệu suất của nó trên tập <code>validation dataset</code>. Để làm điều này, chúng ta có thể sử dụng phương thức <code>evaluate</code>.</p>

````python
# 1. Prepare data for evaluation
raw_val_data = val['dialogue'].tolist()  # Get dialogues as a list
raw_val_labels = val['summary'].tolist()  # Get summaries (labels) as a list

# 2. Generate predictions
def generate_summary(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True).to(device)
    summary_ids = model.generate(**inputs)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

predictions = [generate_summary(text) for text in raw_val_data]

# 3. Create a Dataset for evaluation
eval_dataset = Dataset.from_dict({'predictions': predictions, 'references': [[label] for label in raw_val_labels]}) # Wrap labels in a list for ROUGE

# 4. Load evaluation metric
metric = evaluate.load("rouge")

# 5. Compute metrics
results = metric.compute(predictions=eval_dataset["predictions"], references=eval_dataset["references"])
print(results)
````

Mô hình tóm tắt văn bản có khả năng chọn ra các từ quan trọng (rouge1: 0.24), nhưng chưa thực sự tốt trong việc tạo ra bản tóm tắt trôi chảy và tự nhiên (rouge2: 0.06, rougeL: 0.19). Vẫn cần cải thiện mô hình để đạt được kết quả tốt hơn.

Điểm số ROUGE dao động từ 0 đến 1, với các giá trị cao hơn cho thấy sự tương đồng lớn hơn giữa bản tóm tắt do mô hình tạo ra và bản tóm tắt tham chiếu.

Cụ thể hơn:

0: Cho biết không có sự trùng lặp nào giữa bản tóm tắt do mô hình tạo ra và bản tóm tắt tham chiếu.
1: Cho biết sự trùng lặp hoàn hảo giữa bản tóm tắt do mô hình tạo ra và bản tóm tắt tham chiếu.
Trong thực tế, điểm số ROUGE hiếm khi đạt đến 1. Điểm số thường nằm trong khoảng từ 0 đến 0.6, tùy thuộc vào độ phức tạp của tác vụ tóm tắt và chất lượng của mô hình.

Ví dụ:

Điểm số ROUGE là 0.4 cho thấy một mức độ tương đồng vừa phải giữa bản tóm tắt do mô hình tạo ra và bản tóm tắt tham chiếu.
Điểm số ROUGE là 0.6 cho thấy một mức độ tương đồng cao, mô hình đang hoạt động khá tốt.
Lưu ý:

Không có một ngưỡng điểm số ROUGE cụ thể nào được coi là "tốt" hay "xấu". Việc giải thích điểm số phụ thuộc vào tác vụ cụ thể và tập dữ liệu được sử dụng.
Bạn nên so sánh điểm số ROUGE của mô hình với điểm số của các mô hình khác hoặc với một baseline (ví dụ: một phương pháp tóm tắt đơn giản) để đánh giá hiệu suất tương đối của mô hình.
